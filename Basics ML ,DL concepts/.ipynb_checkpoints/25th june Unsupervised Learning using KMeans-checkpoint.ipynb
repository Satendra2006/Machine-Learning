#### Step-1: Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
!pip install xlrd
```

#### Step-1: Import Dataset

```python
dataset=pd.read_excel('Customers_ID.xls')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

#### Step-3: To form feature matrix(X)

```python
X=dataset.iloc[:,1:].values
```

#### Step-4: Visualize the feature matrix

```python
plt.scatter(X[:,0],X[:,1],c='red')
plt.show()
```

#### Step-5: Building clustering model using KMeans

```python
from sklearn.cluster import KMeans
```

```python
KM=KMeans(n_clusters=4)
```

```python
### Testing the model
KM.fit(X)
```

```python
# Testing the model:
Ypred_Kmeans=KM.predict(X)
```

```python
Ypred_Kmeans
```

```python
plt.scatter(X[Ypred_Kmeans==0,0],X[Ypred_Kmeans==0,1],c='blue',label='Cluster 0')
plt.scatter(X[Ypred_Kmeans==1,0],X[Ypred_Kmeans==1,1],c='red',label='Cluster 1')
plt.scatter(X[Ypred_Kmeans==2,0],X[Ypred_Kmeans==2,1],c='yellow',label='Cluster 2')
plt.scatter(X[Ypred_Kmeans==3,0],X[Ypred_Kmeans==3,1],c='green',label='Cluster 3')
plt.legend()
plt.show()
```

#### Choosing optimal value of K

```python
errorrate=[]
for i in range(1,31):
    KM=KMeans(n_clusters=i)
    KM.fit(X)
    errorrate.append(KM.inertia_)
```

```python
plt.plot(range(1,31),errorrate,marker='o',markerfacecolor='red',markersize=5)
plt.xlabel('Clusters (K)')
plt.yalbel('Inertia')
plt.show()
```
