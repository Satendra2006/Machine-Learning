#### Step-1: Generate Dataset:

```python
# !pip install pandas
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.linspace(-3,3,1000).reshape(-1,1)
Y=X**2
```

```python
plt.scatter(X,Y)
plt.show()
```

#### Step-2: Split the dataset as training dataset and testing dataset

```python
# !pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
```

#### Step-3: Building Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_model=LinearRegression()
```

```python
## Train the model
lin_model.fit(xtrain,ytrain)
# Test the model
ypred_lin=lin_model.predict(xtest)
```

```python
# Visualise the predicted model and the actual model
plt.scatter(xtest,ytest,label='Actual Data')
plt.plot(xtest,ypred_lin,label='Predicted Data',c='orange')
plt.legend()
plt.show()
```

#### Step-4: Building with ANN Model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Architecture of ANN
nl_model=Sequential()
layer=Dense(20,activation='relu',input_shape=(1,))
nl_model.add(layer) ###Hiden Layer
nl_model.add(Dense(20,activation='relu'))
nl_model.add(Dense(1))
```

```python
# Compile ANN
nl_model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
```

```python
# Train the model
nl_model.fit(xtrain,ytrain,epochs=100,verbose=0)
```

```python
# Test the model:
ypred_nl=nl_model.predict(xtest)
```

```python
# Visualize the model
plt.scatter(xtest,ypred_nl,c='red',label='Predicted NN Model')
plt.scatter(xtest,ytest,label='Actual Model')
plt.plot(xtest,ypred_lin,label='Predicted Linear model',c='green')
plt.legend()
plt.show()
```

#### Evaluate ANN model using some metrics

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse=mean_squared_error(ytest,ypred_nl)
mae=mean_absolute_error(ytest,ypred_nl)
r2=r2_score(ytest,ypred_nl)
```

```python
print("MSE:{}".format(mse))
print("MEA:{}".format(mae))
print("R2 Score:{}".format(r2))
```
